```python
import matplotlib.pyplot as plt
x =[1,2,3,4,5]
y =[2,3,1,4,5]
# ms=10: Sets the marker size to 10.
# mfc='green': Sets the marker face color to green.
plt.step(x,y, marker='o' , color= 'red', ms =10, mfc = 'green' , label='struggle')
# plt.legend(loc='upper left'): Adds a legend to the plot at the upper left corner.
plt.legend(loc='upper left')
plt.grid()
plt.title('Language')
plt.xlabel('Python')
plt.ylabel('Score')
plt.show()
```
