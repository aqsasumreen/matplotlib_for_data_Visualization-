```python
import matplotlib.pyplot as plt
import numpy as np
x = ["python", "c++", "JS", "R"]
y= [10,20,30,40]
z = [5,15,25,35]
width=0.2
p =np.arange(len(x))
p1 = [j + width for j in p] # # One value is its own, and adding the second width will move it forward, meaning that one width will be here, and the second will just come next to it.

plt.xlabel("language", fontsize = 15)
plt.ylabel("numbers")
plt.title("survey")
col = ['r', 'b', 'y', 'purple']
plt.bar(p, y, width, color=col,  edgecolor="orange" , label="popularity1")
plt.bar(p1, z, width, color='g', edgecolor="orange" , alpha=0.5, label="popularity2")
plt.xticks(p + width/2, x, rotation= 5) # xticks used to show names
plt.legend()
plt.show()
```
