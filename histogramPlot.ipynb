```python
no =  [38, 21, 44, 42, 31, 43 ,43, 24, 26, 52, 39, 20, 12 ,45, 49, 50, 24, 25, 35, 39, 13, 27, 48, 29, 20, 11, 39, 40, 56, 23, 30, 21, 11, 31, 50, 32, 18, 22, 50, 48, 12, 47, 28, 41, 44, 27, 16 ,10,28, 33]
# between 10-20 a half bar , similary between 20-30 ...... this can be solved by bins parameter bins = l:
l =[10, 20, 30, 40, 50] # to increase range add (0,100) in hist
#  to reverse the order of graph add cumulative = -1
# bottom =10 , bottom 10 se start hoga
# histtype = bar , barstack, step, filled
#  to convert in horizental , orientation = horizontal,
# align = 'left'
# , rwidth=0.8 
# TO CONVERT into log.
plt.hist(no, edgecolor = 'r',bins = l, bottom = 10 , log=True , label = "langgg")
plt.title('histogram')
plt.xlabel('language')
plt.ylabel('numbers')
plt.axvline(35 , color="green" , label="line")
# You can use plot or vlines to draw a vertical line, but to draw a vertical line from the bottom to the top of the y axis, axvline is the probably the simplest function to use.
plt.legend()
plt.grid()
plt.show()
```
