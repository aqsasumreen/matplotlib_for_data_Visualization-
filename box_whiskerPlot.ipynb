```python
# whisker is the upper and lower part of box plot , used to show maximum and minimum

import matplotlib.pyplot as plt
x =[10, 20, 30, 40, 50, 60, 70, 120]
x1=[10,20,30,35,40,50,60,70]
# to tilit the plt add  notch=True
#  for horixontol plot add  vert = False
#  Add whis to join with outiler ,discoonect when it is false
# sym='g+' -- chg the color and shape of outlier


plt.boxplot(x, vert = False , widths= 0.4, labels=['Python'], patch_artist= True,sym='g+', boxprops=dict(color = 'yellow'), capprops=dict(color = 'purple'), showmeans=True,whiskerprops=dict(color = 'green'))
# to add new plot
plt.show()
```

```python
# to add new plot
l =[x, x1]
plt.boxplot(l, vert = False ,labels=["Python", "JS"], widths= 0.4, patch_artist= True,sym='g+', boxprops=dict(color = 'yellow'), capprops=dict(color = 'purple'), whiskerprops=dict(color = 'green'))
plt.show()
```
