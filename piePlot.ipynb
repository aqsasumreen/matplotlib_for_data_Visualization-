```python
# Most important in python development
import matplotlib.pyplot as plt
x =[10, 20, 30,40]
y = ['c++', 'php', 'java','python']
# We want the C++ part to be a bit separated; for this, we use explode.
# To show percentage values, use autopct, where 0.1 means one value after the decimal point.
# labeldistance = to set the distance of the label parameter.
# startangle = 0, 90, 180, 270, 360 to set the starting angle.
# counterclockwise and clockwise movement.
# center = (1, 3) used to change the center.
# radius = 1.3 to increase the radius of the plot.

ex = [0.3, 0.0, 0.0, 0.0]
plt.pie(x, labels=y , explode= ex, autopct='%0.1f%%', shadow=True , labeldistance = 1.2, startangle = 180, textprops={"fontsize" : 10, "color": 'red'}, counterclock= False, wedgeprops={ "linewidth": 4, "edgecolor": 'm'}, rotatelabels= True)
plt.title("Pie Chart")
plt.legend(loc=4)
plt.show()
```
