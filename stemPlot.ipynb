```python
import matplotlib.pyplot as plt
a = [2,4,6,8,10]
b = [1,3,5,7,9]
## linefmt = ':', '-.', '--', '-' sets the line style.
# markerfmt = 'r*' sets the marker style with color and shape.
# bottom = 9 starts from 9 and the graph goes downward.
# basefmt = 'g' changes the color of the bottom line.
# use_line_collection: False, changes the color based on the baseline.


plt.stem(a, b, linefmt='--', markerfmt='r*', bottom= 0, basefmt= 'g' , label='PYTHON', orientation= 'horizontal')
plt.legend()
plt.show()
```
