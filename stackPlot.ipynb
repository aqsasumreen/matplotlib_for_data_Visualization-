```python
# Stack plot also called area plot
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
a1 =[2,3,4,5,2]
a2 =[1,3,5,4,2]
a3 =[4,3,4,2,1]
l =['area1', 'area2', 'area3']
#  baseline= 'sym' - chg the baseline
plt.stackplot(x, a1, a2, a3,labels=l, baseline= 'sym')
plt.xlabel('Area')
plt.ylabel('height')
plt.legend()
plt.show()
```
