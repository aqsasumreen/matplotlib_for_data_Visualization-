```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,6,7,1,4,5]
z = [3,4,5,6,7,8,9]
size =[413,415,417,519,511,616,618]
# cmap is used for color mapping, colors in specific range of numbers
# 'viridis', 'plasma', 'inferno', 'magma', 'cividis', BrBG
colors =[10,12,20,38,35,55, 45]
plt.scatter(x, y, s = size , marker ='*', edgecolors='green',c =colors, cmap = "BrBG")
# second plot
plt.scatter(x, z, s = size ,c='green' )
t = plt.colorbar()
t.set_label('coloBar' ,  fontsize=10)
plt.title("Scatter Plot", fontsize=10 )
plt.xlabel("Day", fontsize=10)
plt.ylabel("Number", fontsize=10) 
plt.show()

```
